```python
# coding: utf-8

import os,sys,pathlib
ROOT_PATH = pathlib.Path.cwd().parent.resolve().as_posix()
sys.path.insert(0,ROOT_PATH)
```

```python
from utils import pickle_store, pickle_load,excel_store
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

from sklearn.metrics import roc_curve, auc

from hbrl_hmc import polynomial_transform,reset_gpu

import tensorflow as tf
```

```python
DATA_DIR    = '/media/data/ijcai_2021/'
```

```python
def get_ROC_curve(directory,file_name,verbose=True,store=True):
    """
    Compute the ROC curve and AUC for the each fold in the experiment.
    
    1. load the experiment data
    
    2. for each of the experiment fold, calculate ROC curve and the AUC, and store the result in excel file
    
    3. store the mean, the max and the min of the AUC v.s. the # of leaves
    """

    # run k-fold cross valildation
    #results = pickle_load(module_name='ijcai_2021',prefix='HBRL_L0',verbose=True)
    results   = pickle_load(directory=directory,file_name=file_name,verbose=True)
    print("")
    n_splits  = len(results['classifier'])

    # initialize dicts
    fpr = dict()        # Frue-Positive Rate
    tpr = dict()        # True-Positive Rate
    roc_auc = dict()    # 

    # initialize dataframe for the statistics
    df  = pd.DataFrame(columns=['fold#','N_train','N_test','K_target','max_rules','min_rules','min_samples','nSamples','nTune','nCores','n_leaves','pred_0','pred_1','micro_avg','file_name'])
    _auc = np.ones([n_splits])
    _nlv = np.ones([n_splits])

    # calculate dicts
    for j in range(n_splits):
        # for each cross-validation fold

        # print experiment conditions
        classifier = results['classifier'][j]
        L0_list = classifier.L0_list
        L1_list = L0_list['L1_rules']
        try:
            L1_cap  = L0_list['L1_cap_']
        except:
            L1_cap  = L0_list['cap']
            
        print("HBRL classifier #{}: elapsed time = {:7.2f} min".format(j,results['time(sec)'][j]/60.0))
        try:
            print("max_rules ={:4d}, min_rules ={:4d}, min_samples ={:4d}".format(classifier.max_rules,classifier.min_rules,classifier.min_samples))
        except:
            print("max_rules ={:4d}, min_samples ={:4d}, real_val = {:5}".format(classifier.max_rules,classifier.min_samples,classifier.real_val))
        for k in range(len(L1_list)):
            q = np.sum(L1_cap[k])
            L1_rule = L1_list[k]
            if "None" not in str(type(L1_rule)):
                trace = L1_rule.MCMC_sampling['trace']
                varname = trace.varnames[0]                   # we only have one RV
                L1_s    = trace.get_values(varname)           # L1_s ∈ [-inf,inf]
                L1_u    = polynomial_transform(L1_s).numpy()  # L1_u ∈ [0,1]
                L1_r    = (L1_u>=L1_rule._xi_).astype(int)    # L1_s ∈ {0,1}
                nSamples = L1_rule.nSamples                   # # of samples in MCMC sampling
                nTUnes   = L1_rule.nTune                      # # of tunes in MCMC sampling
                nChains  = L1_rule.nCores                     # # of chains in MCMC sampling
                nTarget  = L1_rule.K_target                   # # of L1 rule candidates to search
                print("HBRL #{}, L1 #{}: captures = {:7.1f}, L1 rules = {}, samples = {}, tune = {}, chains = {}".format(j,k,q,nTarget,nSamples,nTUnes,nChains))
            else:
                print("HBRL #{}, L1 #{}: captures = {:7.1f}, default rule\n".format(j,k,q))
        
        # initialize parameters
        n_classes = classifier.L
        try:
            y_prob    = results['y_prob_'][j]
        except:
            y_prob    = results['y_prob'][j]
        y_test    = results['y_test'][j]

        # normalize y_prob ... this shouldn't be necessary...
        y_prob = tf.linalg.normalize(y_prob,ord=1,axis=1)[0].numpy()

        # initialize dicts
        fpr[j] = dict()
        tpr[j] = dict()
        roc_auc[j] = dict()

        for i in range(n_classes):
            fpr[j][i], tpr[j][i], _ = roc_curve(y_true=y_test[:, i], y_score=y_prob[:, i],drop_intermediate=True)
            roc_auc[j][i] = auc(fpr[j][i], tpr[j][i])

        # Compute micro-average ROC curve and ROC area
        # numpy.ravel(): return a contiguous flattened array.
        fpr[j]["micro"], tpr[j]["micro"], _ = roc_curve(y_test.ravel(), y_prob.ravel())
        roc_auc[j]["micro"] = auc(fpr[j]["micro"], tpr[j]["micro"])

        model    = results['classifier'][j]
        L0_list  = model.L0_list
        L1_list  = L0_list['L1_rules']
        n_leaves = 0
        for k in range(len(L1_list)):
            L1_rule = L1_list[k]
            if "None" not in str(type(L1_rule)): n_leaves += np.sum(L1_rule.L1_r)
            else                               : n_leaves += 1   # default rule
        L1_rule = L1_list[0]
        N_train = L1_rule.dim0[1]
        N_test  = results['y_test'][j].shape[0]

        row = {}
        row['fold#']       = j    
        try:
            row['N_train'] = model.N_train
            row['N_test']  = model.N_test
        except:
            row['N_train'] = N_train
            row['N_test']  = N_test
        row['max_rules']   = model.max_rules    
        row['min_samples'] = model.min_samples
        try:
            row['min_rules'] = model.min_rules
        except:
            row['min_rules'] = None
        try:
            if model.real_val: row['real_val'] = True
            else             : row['real_val'] = False
        except:
            row['real_val'] = None
        row['nSamples']    = L1_rule.nSamples    
        row['nTune']       = L1_rule.nTune    
        row['nCores']      = L1_rule.nCores    
        row['K_target']    = L1_rule.K_target    
        row['n_leaves']    = _nlv[j] = n_leaves
        row['pred_0']      = roc_auc[j][0]    
        row['pred_1']      = _auc[j] = roc_auc[j][1]    
        row['micro_avg']   = roc_auc[j]['micro']    
        row['file_name']   = file_name
        df = df.append(row,ignore_index=True)

    if not store:
        # do not append the data to the statistics file
        return fpr,tpr,roc_auc,n_splits,results
        
    file_name_excel = "HBRL_stats.xlsx"
    sheet_name      = "exp"
    if excel_store(df,directory=DATA_DIR,file_name=file_name_excel,sheet_name=sheet_name,add_sheet=True,append=True):
        print('statistics successfully wrote to file: {}{}.'.format(DATA_DIR,file_name_excel))

    # append the summary
    (auc_max,auc_min,auc_avg) = (np.amax(_auc),np.amin(_auc),np.mean(_auc))
    (nlv_max,nlv_min,nlv_avg) = (np.amax(_nlv),np.amin(_nlv),np.mean(_nlv))
    N = N_train + N_test
    sheet_name = 'summary'
    dfs = pd.DataFrame({'auc_max':[auc_max],'auc_min':[auc_min],'auc_avg':[auc_avg],'nlv_max':[nlv_max],'nlv_min':[nlv_min],'nlv_avg':[nlv_avg],'N':N,'file_name':file_name})
    if excel_store(dfs,directory=DATA_DIR,file_name=file_name_excel,sheet_name=sheet_name,add_sheet=True,append=True):
        print('statistics successfully appended to file={}{} : worksheet={}.'.format(DATA_DIR,file_name_excel,sheet_name))
        
    print("\n---end of {}\n".format(directory + file_name))
    
    return fpr,tpr,roc_auc,n_splits,results
```

```python
# scan all the LGBM result files and store statistics
list_dir = os.listdir(DATA_DIR)
for file_name in list_dir:
    if "HBRL_N=1000" in file_name:
        fpr,tpr,roc_auc,n_splits,results = get_ROC_curve(directory=DATA_DIR,file_name=file_name)
```

## Step #1: Load the HBRL-L0 results and calculate ROC curve

```python
# run k-fold cross valildation
file_name = 'HBRL_N=1000_K=5_R=3_U=0_M=10_S=50_T=50_2021-01-18_21:59:10.pkl'
fpr,tpr,roc_auc,n_splits,results = get_ROC_curve(directory=DATA_DIR,file_name=file_name,store=False)
```

## Step #2: Show ROC curve

```python
lw = 2

for j in range(n_splits):

    # rule vector size distribution
    plt.figure()

    plt.subplot(211)
    plt.plot(fpr[j][0], tpr[j][0], color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[j][0])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic: fold #{}'.format(j))
    plt.legend(loc="lower right")

    plt.subplot(212)
    plt.plot(fpr[j][1], tpr[j][1], color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[j][1])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic: fold #{}'.format(j))
    plt.legend(loc="lower right")

    plt.subplots_adjust(top=4.00, bottom=0.10, left=0.10, right=1.40, hspace=0.13, wspace=1.00)

    plt.show()
```
