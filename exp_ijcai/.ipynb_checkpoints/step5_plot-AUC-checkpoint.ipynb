```python
# coding: utf-8

import os,sys,pathlib
ROOT_PATH = pathlib.Path.cwd().parent.resolve().as_posix()
sys.path.insert(0,ROOT_PATH)
```

```python
import numpy as np
import pandas as pd

from utils import pickle_store, pickle_load, excel_store, excel_load
import matplotlib.pyplot as plt
```

```python
DATA_DIR    = '/media/data/ijcai_2021/'
```

```python
# run k-fold cross valildation
#results = pickle_load(module_name='ijcai_2021',prefix='HBRL_L0',verbose=True)
file_name = 'HBRL_stats.xlsx'
worksheets   = excel_load(directory=DATA_DIR,module_name='ijcai_2021',file_name=file_name,verbose=True)
```

```python
def get_statistics(directory,file_name):
    
    # load dataframe from file
    worksheets = excel_load(directory=DATA_DIR,file_name=file_name,verbose=True)
    dfs        = worksheets['summary']
    
    # take sample points and their errors
    n_points   = dfs.shape[0]
    x          = dfs['nlv_avg'].to_numpy()
    y          = dfs['auc_avg'].to_numpy()
    xerr_n     = x - dfs['nlv_min'].to_numpy()
    xerr_p     = dfs['nlv_max'].to_numpy() - x
    xerr       = np.concatenate((xerr_n.reshape([1,n_points]),xerr_p.reshape([1,n_points])),axis=0)
    yerr_n     = y - dfs['auc_min'].to_numpy()
    yerr_p     = dfs['auc_max'].to_numpy() - y
    yerr       = np.concatenate((yerr_n.reshape([1,n_points]),yerr_p.reshape([1,n_points])),axis=0)

    return x,y,xerr,yerr
```

```python
x,y,xerr,yerr = get_statistics(directory=DATA_DIR,file_name='HBRL_stats.xlsx')
```

```python
# benchmark with simple plots

plt.figure()

plt.subplot(211)

x,y,xerr,yerr = get_statistics(directory=DATA_DIR,file_name='HBRL_stats.xlsx')
plt.scatter(x, y, marker='^',color='r',s=100,label='HBRL')

print('\n\n')

plt.xlabel('# of leaves', fontsize=20)
plt.ylabel('AUC ROC', fontsize=20)
plt.title('',fontsize=20)
plt.legend(loc="lower right", fontsize=20)
plt.xlim(3.0,20.0)
plt.ylim(0.5,1.0)
plt.xscale("log")

plt.subplots_adjust(top=4.00, bottom=0.10, left=0.10, right=1.40, hspace=0.13, wspace=1.00)

plt.show()
```

```python
# benchmark with error bars

plt.figure()

plt.subplot(211)

x,y,xerr,yerr = get_statistics(directory=DATA_DIR,file_name='HBRL_stats.xlsx')
plt.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='o',color='r',)

print('\n\n')

plt.xlabel('n_leaves')
plt.ylabel('ROC Area')
plt.title('HBRL result: ROC Area v.s. # of leaves')
plt.xlim(3.0,20.0)
plt.ylim(0.5,1.0)
plt.xscale("log")

plt.subplots_adjust(top=4.00, bottom=0.10, left=0.10, right=1.40, hspace=0.13, wspace=1.00)

plt.show()
```
